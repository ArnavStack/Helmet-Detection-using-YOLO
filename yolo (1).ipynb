```python
from ultralytics import YOLO

# Load a pretrained YOLOv8 model
model = YOLO("yolov8n.pt")
```

```python
# Train YOLOv8 on your dataset
model.train(
    data="/content/helmet_dataset/data.yaml",   # path to your dataset yaml
    epochs=50,                                  # number of epochs (you can increase)
    imgsz=640                                   # image size
)
```

```python
# Load the trained model
model = YOLO("/content/runs/detect/train/weights/best.pt")

# Run prediction on one image
results = model.predict(
    source="/content/helmet_dataset/valid/images/your_test_image.jpg",  # change file name
    show=True,   # display inside Colab
    save=True    # save output
)
```
